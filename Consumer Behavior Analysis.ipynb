#  Scaling

```python
#Rule of thumb: Use StandardScaler for normally distributed data, otherwise use MinMaxScaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stdscaler=StandardScaler()
```

```python
age = data['Age'].values.reshape(-1, 1)
```

```python
scaled_feature = stdscaler.fit_transform(age) 
data['Age'] = scaled_feature
```

```python
#AnnualIncome
shaped_income=data['AnnualIncome'].values.reshape(-1,1)
scaled_income=stdscaler.fit_transform(shaped_income)
data['AnnualIncome']=scaled_income
```

```python
#TimeSpentOnWebsite
shaped_websiteTime=data['TimeSpentOnWebsite'].values.reshape(-1,1)
scaled_websiteTime=stdscaler.fit_transform(shaped_websiteTime)
data['TimeSpentOnWebsite']=scaled_websiteTime
```

```python
#NumberOfPurchases:
shaped_purchases=data['NumberOfPurchases'].values.reshape(-1,1)
scaled_purchases=stdscaler.fit_transform(shaped_purchases)
data['NumberOfPurchases']=scaled_purchases
```

#  Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(data.drop('PurchaseStatus',axis=1),data[['PurchaseStatus']],test_size=0.2,random_state=1)
```

#  Model Selection

```python
#Knn & logistic regression:
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
knn=KNeighborsClassifier()
lr=LogisticRegression()
```

```python
#KNN Prediction:
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
```

```python
#Logistic Prediction:
lr.fit(X_train,y_train)
```

```python
y_pred_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
#For KNN:
accuracy_knn=accuracy_score(y_test,y_pred_knn)
print('Accuracy Score of KNN algo:',accuracy_knn)
```

```python
#For LR:
accuracy_lr=accuracy_score(y_test,y_pred_lr)
print('Accuracy Score of Logistic Regression algo:',accuracy_lr)
```

```python
#XGBoost:
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
xgb.fit(X_train,y_train)
```

```python
y_pred_xgb=xgb.predict(X_test)
```

```python
#For XGB:
accuracy_xgb=accuracy_score(y_test,y_pred_xgb)
print('Accuracy Score of XGBoost algo:',accuracy_xgb)
```

```python
#Random Forest Classifier:
from sklearn.ensemble import RandomForestClassifier
```

```python
random=RandomForestClassifier()
```

```python
random.fit(X_train,y_train)
```

```python
y_pred_random=random.predict(X_test)
```

```python
accuracy_random=accuracy_score(y_test,y_pred_random)
print('Accuracy Score of Random forest algorithm algo:',(accuracy_random)*100)
```

```python
#Support Vector Machine:
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred_svc=svc.predict(X_test)
```

```python
accuracy_svc=accuracy_score(y_test,y_pred_svc)
print('Accuracy Score of SVC algo:',accuracy_svc)
```

```python
#Proba Score:
randomForest_proba=random.predict_proba(X_test)
randomForest_proba
```

```python
#Decision function:
y_train_proba_random=random.predict_proba(X_train)
y_test_proba_random=random.predict_proba(X_test)
```

```python
#AUC AND ROC CURVE [TRAIN DATASET]:
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
r_fpr,r_tpr,threshold=roc_curve(y_train,y_train_proba_random[:, 1])
auc_r=auc(r_fpr,r_tpr)
plt.plot(r_fpr, r_tpr, linestyle='-',label= 'RandomForest (auc=%0.3f)' %auc_r)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```

```python
#AUC AND ROC CURVE [TEST DATASET]
rt_fpr,rt_tpr,threshold=roc_curve(y_test,y_test_proba_random[:, 1])
auc_tr=auc(rt_fpr,rt_tpr)
plt.plot(rt_fpr, rt_tpr, linestyle='-',label= 'RandomForest (auc=%0.3f)' %auc_tr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
